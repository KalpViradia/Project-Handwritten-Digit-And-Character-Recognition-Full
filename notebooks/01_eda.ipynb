# Handwritten Digit Recognition - Exploratory Data Analysis

This notebook explores the MNIST dataset and demonstrates the preprocessing pipeline.

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow.keras.datasets import mnist

# Set style
plt.style.use('dark_background')
sns.set_palette('husl')
```

## 1. Load MNIST Dataset

```python
# Load data
(X_train, y_train), (X_test, y_test) = mnist.load_data()

print(f'Training samples: {X_train.shape[0]}')
print(f'Test samples: {X_test.shape[0]}')
print(f'Image shape: {X_train.shape[1:]} (28x28 grayscale)')
print(f'Classes: {np.unique(y_train)}')
```

## 2. Visualize Sample Digits

```python
# Display sample images
fig, axes = plt.subplots(3, 10, figsize=(15, 5))
fig.suptitle('Sample MNIST Digits', fontsize=16)

for i, ax in enumerate(axes.flat):
    ax.imshow(X_train[i], cmap='gray')
    ax.set_title(f'{y_train[i]}')
    ax.axis('off')

plt.tight_layout()
plt.show()
```

## 3. Class Distribution

```python
# Plot class distribution
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Training set
unique, counts = np.unique(y_train, return_counts=True)
ax1.bar(unique, counts, color=plt.cm.viridis(np.linspace(0, 0.8, 10)))
ax1.set_xlabel('Digit')
ax1.set_ylabel('Count')
ax1.set_title('Training Set Distribution')
ax1.set_xticks(range(10))

# Test set
unique, counts = np.unique(y_test, return_counts=True)
ax2.bar(unique, counts, color=plt.cm.viridis(np.linspace(0, 0.8, 10)))
ax2.set_xlabel('Digit')
ax2.set_ylabel('Count')
ax2.set_title('Test Set Distribution')
ax2.set_xticks(range(10))

plt.tight_layout()
plt.show()
```

## 4. Pixel Value Distribution

```python
# Analyze pixel values
print(f'Pixel value range: {X_train.min()} - {X_train.max()}')
print(f'Mean pixel value: {X_train.mean():.2f}')
print(f'Std pixel value: {X_train.std():.2f}')

plt.figure(figsize=(10, 4))
plt.hist(X_train.flatten(), bins=50, color='cyan', alpha=0.7)
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.title('Pixel Value Distribution (Before Normalization)')
plt.show()
```

## 5. Preprocessing Steps

```python
from tensorflow.keras.utils import to_categorical

# Normalize to [0, 1]
X_train_norm = X_train.astype('float32') / 255.0
X_test_norm = X_test.astype('float32') / 255.0

print(f'After normalization: {X_train_norm.min():.2f} - {X_train_norm.max():.2f}')

# Reshape for CNN
X_train_reshaped = X_train_norm.reshape(-1, 28, 28, 1)
X_test_reshaped = X_test_norm.reshape(-1, 28, 28, 1)

print(f'Shape for CNN: {X_train_reshaped.shape}')

# One-hot encode labels
y_train_cat = to_categorical(y_train, 10)
y_test_cat = to_categorical(y_test, 10)

print(f'One-hot label example: {y_train[0]} -> {y_train_cat[0]}')
```

## 6. Visualize One Sample Per Class

```python
# Show one sample per digit
fig, axes = plt.subplots(2, 5, figsize=(15, 6))
fig.suptitle('One Sample Per Digit Class', fontsize=16)

for digit in range(10):
    idx = np.where(y_train == digit)[0][0]
    ax = axes[digit // 5, digit % 5]
    ax.imshow(X_train[idx], cmap='gray')
    ax.set_title(f'Digit: {digit}', fontsize=12)
    ax.axis('off')

plt.tight_layout()
plt.show()
```

## 7. Mean Image Per Class

```python
# Calculate mean image for each class
fig, axes = plt.subplots(2, 5, figsize=(15, 6))
fig.suptitle('Mean Image Per Digit Class', fontsize=16)

for digit in range(10):
    mask = y_train == digit
    mean_img = X_train[mask].mean(axis=0)
    
    ax = axes[digit // 5, digit % 5]
    ax.imshow(mean_img, cmap='hot')
    ax.set_title(f'Digit: {digit}', fontsize=12)
    ax.axis('off')

plt.tight_layout()
plt.show()
```

## Summary

- MNIST has 70,000 28x28 grayscale images
- Dataset is well-balanced across all 10 classes
- Preprocessing: Normalization, reshaping, one-hot encoding
- Ready for model training!
